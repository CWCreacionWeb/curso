# Instrucciones para Resolver Sistemas Ax=b y By=b

Se te proporciona una serie de tablas en la cual la 칰ltima columna corresponde a un elemento conocido como **etiqueta**, y el resto de columnas se conoce como **caracter칤sticas**. ### 1. Descomponer las tablas
Descomp칩n las tablas en:
- **Matriz A:** que contiene todas las caracter칤sticas.
- **Vector b:** que contiene las etiquetas.

### 2. Generar la Matriz Ampliada B
Genera una matriz ampliada **B** a침adiendo una primera columna de $1's$ y el resto con los valores de la matriz **A**, de modo que:
$$ B = [1|A] $$

### 3. C치lculo del Determinante
En los casos en que sea posible, calcula el determinante de las matrices **A** y **B**:
- Determinante de **A**: $ \text{det}(A) $
- Determinante de **B**: $ \text{det}(B) $

### 4. C치lculo del Rango
Calcula el rango de las matrices:
- **Rango de A**: $ \text{rank}(A) $
- **Rango de B**: $ \text{rank}(B) $

Tambi칠n, calcula el rango de las matrices ampliadas **[A|b]** y **[B|b]**:
$$ \text{rank}( [A|b] ) \quad \text{y} \quad \text{rank}( [B|b] ) $$

### 5. Clasificaci칩n de los Sistemas
Clasifica los sistemas:
- **Ax=b**
- **By=b**

seg칰n la clasificaci칩n de sistemas de ecuaciones lineales (determinado, indeterminado, o incompatible) basado en los rangos obtenidos.

### 6. Resoluci칩n de los Sistemas
Resuelve los sistemas utilizando:
- **Matriz inversa** si el sistema es determinado: $ x = A^{-1}b $
- **Pseudoinversa** en caso de ser un sistema indeterminado: $ x = A^{+}b $

Haz lo mismo para **B**:
$$ y = B^{-1}b \quad \text{o} \quad y = B^{+}b $$

### 7. An치lisis de Estabilidad
Realiza un an치lisis de estabilidad de los sistemas a trav칠s de dilataciones y contracciones de las soluciones en factores del:
- $0.5\%$, $1\%$, $2\%$, $5\%$, y $10\%$.

Analiza los cambios porcentuales que se generan en las soluciones $b'$ perturbadas en relaci칩n con la $b$ real.

### 8. Intervalos de Confianza
Con el an치lisis anterior, establece una clase de intervalos de confianza en terminos de porcentaje para las soluciones obtenidas, considerando que:
- Las **dilataciones** podr칤an simular el redondeo o p칠rdida de precisi칩n en los decimales de las soluciones.
- Las **contracciones** podr칤an simular errores num칠ricos peque침os o fluctuaciones en los datos originales.

### 9. Interpolantes
Las soluciones de los sistemas **Ax=b** y **By=b** pueden utilizarse para construir interpolantes:
- **Sin t칠rmino independiente** para **Ax=b**.
- **Con t칠rmino independiente** para **By=b**.

Construye dichos interpolantes.

```python
import numpy as np
# tabla (6x6)
tabla = np.array([[0.32047, 0.82356, 0.25645, 0.49580, 0.67355, 0.61427],
                  [0.29187, 0.44708, 0.19897, 0.89325, 0.73233, 0.89347],
                  [0.13883, 0.64831, 0.02541, 0.84798, 0.57868, 0.37958],
                  [0.89019, 0.36956, 0.23375, 0.45624, 0.18749, 0.25037],
                  [0.54278, 0.18745, 0.96885, 0.25942, 0.65492, 0.12485],
                  [0.76956, 0.06538, 0.49874, 0.82947, 0.86217, 0.78432]])

# Paso 1: Descomponer la tabla
# Descompondremos la tabla en la matriz A (todas las columnas menos la 칰ltima) y el vector b (la 칰ltima columna).
A = tabla[0:,:-1] # Matriz A (todas las columnas menos la 칰ltima)


# Vector b (칰ltima columna)
b = tabla[0:, -1] #selecciona todas las filas : y solo la 칰ltima columna -1.

# Paso 2: Generar la Matriz Ampliada B
# La matriz ampliada 洧냣
# B se obtiene a침adiendo una columna inicial de unos a la matriz 洧냢
# A. Esto se puede hacer as칤:
# A침adir una columna de unos a la matriz A para formar la matriz ampliada B
B = np.hstack([np.ones((A.shape[0], 1)), A])


# Verificar si la matriz A es cuadrada antes de calcular su determinante
if A.shape[0] == A.shape[1]:
    det_A = np.linalg.det(A)
    print("Determinante de A:", det_A)
else:
    print("La matriz A no es cuadrada, no se puede calcular el determinante.")

# Verificar si la matriz B es cuadrada antes de calcular su determinante
if B.shape[0] == B.shape[1]:
    det_B = np.linalg.det(B)
    print("Determinante de B:", det_B)
else:
    print("La matriz B no es cuadrada, no se puede calcular el determinante.")


# Calculo del rango
rango_A = np.linalg.matrix_rank(A)
print(f"El rango de A es {rango_A}")
# Verificaci칩n con el n칰mero de columnas
num_columnas_A = A.shape[1]
if rango_A == num_columnas_A:
    print("Todas las caracter칤sticas son linealmente independientes.")
else:
    print("Hay caracter칤sticas linealmente dependientes.")

```
